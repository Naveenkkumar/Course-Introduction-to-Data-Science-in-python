```python
def Clean_names(Country_name):
    if re.search('\(.*', Country_name):
        pos = re.search('[ ]\(.*', Country_name).start()
        return Country_name[:pos]
    else:
        return Country_name
    

##
def Clean_names_numbers(Country_name):
    if re.search('[0-9].*$', Country_name):
        pos = re.search('[0-9].*$', Country_name).start()
        return Country_name[:pos]
    else:
        return Country_name
    

    


def answer_one():
    # YOUR CODE HERE
    Energy = pd.read_excel(r'C:\\Users\\Naveen Kumar\\OneDrive - RxLogix\\Documents\\PersoMe\\MLcodes\\Courses\\Intro to data science in python\\assignments\\assignment3\\assets\\Energy Indicators.xls',
                      header = None)
    Energy = Energy.iloc[18:245,2:]
    Energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
    Energy.reset_index(drop = True, inplace = True)
    Energy['Energy Supply'].replace(to_replace = '...', value = np.nan, inplace = True)
    Energy['Energy Supply'] = [x*1000000 for x in Energy['Energy Supply']]
    Energy.replace({'Country':{"Republic of Korea": "South Korea", "United States of America20": "United States",
                               "United Kingdom of Great Britain and Northern Ireland19": "United Kingdom",
                               "China, Hong Kong Special Administrative Region3": "Hong Kong"}}, inplace = True)
    
    Energy['Country'] = Energy['Country'].apply(Clean_names)
    Energy['Country'] = Energy['Country'].apply(Clean_names_numbers)
    #print(Energy)
    
    ###
    GDP = pd.read_csv('C:\\Users\\Naveen Kumar\\OneDrive - RxLogix\\Documents\\PersoMe\\MLcodes\\Courses\\Intro to data science in python\\assignments\\assignment3\\assets\\world_bank.csv',
                     header = None)
    #GDP = GDP.iloc[4:,:]
    GDP, GDP.columns = GDP.iloc[5:,:], GDP.iloc[4]
    GDP.reset_index(drop = True, inplace =True)
    GDP.replace({'Country Name':{"Korea, Rep.": "South Korea", "Iran, Islamic Rep.": "Iran",
                                 "Hong Kong SAR, China": "Hong Kong"}}, inplace = True)
    GDP.rename(columns = {'Country Name': 'Country',2006.0:'2006',2007.0:'2007',2008.0:'2008',2009.0:'2009',
                          2010.0:'2010',2011.0:'2011',2012.0:'2012',2013.0:'2013',2014.0:'2014',2015.0:'2015'}, inplace = True)
    GDP_new = GDP[['Country','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']]
    #print(GDP_new)
    
    ScimEn = pd.read_excel(r'C:\\Users\\Naveen Kumar\\OneDrive - RxLogix\\Documents\\PersoMe\\MLcodes\\Courses\\Intro to data science in python\\assignments\\assignment3\\assets\\scimagojr-3.xlsx',
                      header = None)
    
    ScimEn, ScimEn.columns = ScimEn.iloc[1:,:], ScimEn.iloc[0]
    ScimEn.reset_index(drop = True, inplace =True)
    ScimEn_new = ScimEn.iloc[:15,:]
    #print(ScimEn_new)
    #print(ScimEn)
    
    combined_df = pd.merge(pd.merge(ScimEn_new, Energy, on = 'Country'),GDP_new, on = 'Country')
    combined_df.set_index("Country", inplace = True)
    #print(combined_df)
    #print(combined_df.dtypes)
    
    
    return combined_df
    

    
    
    
    

    
    
    
    #raise NotImplementedError()
    
answer_one()
```

```python
def cut_renew(value):
    if value <=10:
        return 'First_Renew_grp'
    elif (value>10) & (value<=20):
        return 'Second_Renew_grp'
    elif (value>20) & (value<=30):
        return 'Third_Renew_grp'
    elif (value>30) & (value<=40):
        return 'Fourth_Renew_grp'
    else:
        return 'Fifth_Renew_grp'
    


def answer_twelve():
    # YOUR CODE HERE
    df = answer_one()
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    
    df['Continent'] = [ContinentDict[x] for x in df.index.values]
    df['Renew_grp'] = df['% Renewable'].apply(cut_renew)
    df.reset_index(inplace = True)
    df_grpp = df.groupby(['Continent','Renew_grp']).size()#agg(size =('Country',len))
    return df_grpp
    #print(df.head())
    
    
    #raise NotImplementedError()
answer_twelve()
```

```python
def to_string(value):
    value = str(value)
    value_splited = value.split('.')
    intt = value_splited[0]
    frac =  value_splited[1]
    if len(intt)%3 == 0:
        j = 0
        intt_new = intt
        for i in range(3,len(intt),3):
            intt_new = intt_new[:i+j]+','+intt_new[i+j:]
            j = j + 1
        pop_string = intt_new + '.' + frac
        return pop_string
    elif len(intt)%3 == 1:
        j = 0
        intt_first = intt[0]+','
        intt_last = intt[1:]
        intt_new = intt_last
        for i in range(3,len(intt_last),3):
            intt_new = intt_new[:i+j]+','+intt_new[i+j:]
            j = j + 1
        pop_string = intt_first + intt_new + '.' + frac
        return pop_string
    else:
        j = 0
        intt_first = intt[:1]+','
        intt_last = intt[2:]
        intt_new = intt_last
        for i in range(3,len(intt_last),3):
            intt_new = intt_new[:i+j]+','+intt_new[i+j:]
            j = j + 1
        pop_string = intt_first + intt_new + '.' + frac
        return pop_string
        
        
        
            
            
    

def answer_thirteen():
    # YOUR CODE HERE
    df = answer_one()
    df['Population_estimate'] = (df['Energy Supply']/df['Energy Supply per Capita']).astype(float)
    df['PopEst'] = df['Population_estimate'].apply(to_string)
    
    return df['PopEst']
    
    #raise NotImplementedError()
answer_thirteen()
```
