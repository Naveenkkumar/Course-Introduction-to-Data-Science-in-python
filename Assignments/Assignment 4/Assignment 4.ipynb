```python
import pandas as pd
import numpy as np
import re
import scipy.stats as stats

def nhl_correlation(): 
    # YOUR CODE HERE
    population_data = pd.read_html('C:\\Users\\Naveen Kumar\\OneDrive - RxLogix\\Documents\\PersoMe\\MLcodes\\Courses\\Intro to data science in python\\assignments\\assignment4\\assets\\wikipedia_data.html')
    population_data = population_data[1]
    population_data = population_data.iloc[:-1,[0,1,2,3,8]]
    population_data['NHL_teams'] = [re.split('\[',x)[0] for x in population_data['NHL']]
    population_data['NHL_teams'].replace('',np.nan, inplace= True)
    population_data['NHL_teams'].replace('—',np.nan, inplace= True)
    population_data.dropna(inplace = True)
    population_data.reset_index(drop = True, inplace = True)
    population_data['Population (2016 est.)[8]'] = pd.to_numeric(population_data['Population (2016 est.)[8]'])
    
    nhl_data = pd.read_csv('C:\\Users\\Naveen Kumar\\OneDrive - RxLogix\\Documents\\PersoMe\\MLcodes\\Courses\\Intro to data science in python\\assignments\\assignment4\\assets\\nhl.csv')
    nhl_data = nhl_data.iloc[1:,[0,1,2,3,-2]] # take only relevant columns
    nhl_data = nhl_data[nhl_data['year'] == 2018] # take only 2018 data
    nhl_data.reset_index(drop = True, inplace = True)
    # to remove some strings from W and L columns
    st = ['Metropolitan Division','Central Division','Pacific Division']
    index_names = nhl_data[nhl_data['team'].isin(st)].index
    nhl_data.drop(index_names, inplace = True)
    nhl_data.reset_index(drop = True, inplace = True)
    # calculate W-l % 
    nhl_data['W'] = nhl_data['W'].astype(int)
    nhl_data['L'] = nhl_data['L'].astype(int)
    nhl_data['W/L%'] = nhl_data['W']/(nhl_data['W']+nhl_data['L'])
    nhl_data['team'] = [x.replace('*','') for x in nhl_data['team']]
    
    nhl_data['team_ville'] = nhl_data['team']
    nhl_data['team_ville'] = nhl_data['team_ville'].map({'Tampa Bay Lightning':'Tampa Bay Area',
     'Boston Bruins':'Boston',
     'Toronto Maple Leafs':'Toronto',
     'Florida Panthers':'Miami–Fort Lauderdale',
     'Detroit Red Wings':'Detroit',
     'Montreal Canadiens':'Montreal',
     'Ottawa Senators':'Ottawa',
     'Buffalo Sabres':'Buffalo',
     'Washington Capitals':'Washington, D.C.',
     'Pittsburgh Penguins':'Pittsburgh',
     'Philadelphia Flyers':'Philadelphia',
     'Columbus Blue Jackets':'Columbus',
     'New Jersey Devils':'New York City',
     'Carolina Hurricanes':'Raleigh',
     'New York Islanders':'New York City',
     'New York Rangers':'New York City',
     'Nashville Predators':'Nashville',
     'Winnipeg Jets':'Winnipeg',
     'Minnesota Wild':'Minneapolis–Saint Paul',
     'Colorado Avalanche':'Denver',
     'St. Louis Blues':'St. Louis',
     'Dallas Stars':'Dallas–Fort Worth',
     'Chicago Blackhawks':'Chicago',
     'Vegas Golden Knights':'Las Vegas',
     'Anaheim Ducks':'Los Angeles',
     'San Jose Sharks':'San Francisco Bay Area',
     'Los Angeles Kings':'Los Angeles',
     'Calgary Flames':'Calgary',
     'Edmonton Oilers':'Edmonton',
     'Vancouver Canucks':'Vancouver',
     'Arizona Coyotes':'Phoenix'})
    #print(len(nhl_data['team_ville']))
    
    df = pd.merge(nhl_data,population_data, left_on= "team_ville", right_on= "Metropolitan area")
    df_grp = df.groupby('Metropolitan area').mean().reset_index()

    
    
    
    
    
    #raise NotImplementedError()
    
    population_by_region = df_grp['Population (2016 est.)[8]'] # pass in metropolitan area population from cities
    win_loss_by_region = df_grp['W/L%'] # pass in win/loss ratio from nhl_df in the same order as cities["Metropolitan area"]
    print(len(population_by_region))
    assert len(population_by_region) == len(win_loss_by_region), "Q1: Your lists must be the same length"
    assert len(population_by_region) == 28, "Q1: There should be 28 teams being analysed for NHL"
    
    return stats.pearsonr(population_by_region, win_loss_by_region)


nhl_correlation()
```
